# Scraping Data from the web + Pandas

```python
#import libraries needed

from bs4 import BeautifulSoup
import requests
```

```python
#connect to the website that contains the data you want to pull

url = 'https://en.wikipedia.org/wiki/List_of_largest_oil_and_gas_companies_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
#search for the table to be extracted
# 1st Method

soup.find_all('table')[0]
```

```python
#2nd Method

soup.find('table', class_ ='wikitable sortable')
```

```python
#put your table in a variable

table = soup.find_all('table')[0]
```

```python
#extract the header of the table

soup.find_all('th')
```

```python
world_titles = soup.find_all('th')
```

```python
#change the headers to a list for better formatting and clean the data to remove unwanted spaces and characters

world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
 import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
column_data = soup.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
